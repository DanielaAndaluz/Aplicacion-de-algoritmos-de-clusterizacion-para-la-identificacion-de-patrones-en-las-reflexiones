```python
import pandas as pd
import numpy as np
import os 
import collections
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib import cm
```

```python
#Leemos etiquetas
X=pd.read_csv("C:/Users/daniela.andaluz/Downloads/X_tfidf_index.csv")[['kmeans', 'agnes','dbscan']]
```

```python
#leemos datos del texto original
os.chdir('C:/Users/daniela.andaluz/Downloads')
DATA_TOK_STOP_STEM=np.load("DATA_TOK_STOP_STEM.npy",allow_pickle = True).tolist()
DATA_TOK_STOP=np.load("DATA_TOK_STOP.npy",allow_pickle = True).tolist()
```

```python
#Definimos función que regresa palabras más frecuentes y su frecuencia  por cluster
def palabras_frec(base,vector_clus,clus_num):
    clus_in=[base[x] for x in np.where(vector_clus == clus_num)[0]]
    datos=[]
    for x in clus_in:
        datos=datos+ x
    counter=collections.Counter(datos)
    df=pd.DataFrame({'Keys':list(counter.keys()),'Values':list(counter.values())})
    df=df.sort_values('Values',ascending=False)
    df=df[df['Keys'] !='alumn']
    df=df[df['Keys'] !='estudi']
    d=df[:10]
    return (d)
```

```python
#------------------------------------------------------Kmeans--------------------------------------------------------
df0km=palabras_frec(DATA_TOK_STOP_STEM,X['kmeans'],0)
df1km=palabras_frec(DATA_TOK_STOP_STEM,X['kmeans'],1)
df2km=palabras_frec(DATA_TOK_STOP_STEM,X['kmeans'],2)
df3km=palabras_frec(DATA_TOK_STOP_STEM,X['kmeans'],3)

# Defining the framework of the visualization 
plt.figure(figsize =(20,20)) 
G = gridspec.GridSpec(2, 2) 
ax1 = plt.subplot(G[0, 0]) 
ax2 = plt.subplot(G[0, 1]) 
ax3 = plt.subplot(G[1, 0]) 
ax4 = plt.subplot(G[1, 1]) 

#definimos paletas de colores
viridis = cm.get_cmap('viridis', 10)
plasma=cm.get_cmap('plasma', 10)
inferno=cm.get_cmap('inferno', 10)
cividis=cm.get_cmap('cividis', 10)

n=12

ax1.bar(df0km['Keys'], df0km['Values'],color=viridis.colors)
ax1.set_xticklabels(df0km['Keys'],rotation=45)
ax1.tick_params(axis='x', labelsize=n)
ax1.set_xlabel('Stemm')
ax1.set_ylabel('Frecuencia')
ax1.set_title('Palabras más usadas cluster 1, k-means')

ax2.bar(df1km['Keys'], df1km['Values'],color=plasma.colors)
ax2.set_xticklabels(df1km['Keys'],rotation=45)
ax2.tick_params(axis='x', labelsize=n)
ax2.set_xlabel('Stemm')
ax2.set_ylabel('Frecuencia')
ax2.set_title('Palabras más usadas cluster 2, k-means')

ax3.bar(df2km['Keys'], df2km['Values'],color=inferno.colors)
ax3.set_xticklabels(df2km['Keys'],rotation=45)
ax3.tick_params(axis='x', labelsize=n)
ax3.set_xlabel('Stemm')
ax3.set_ylabel('Frecuencia')
ax3.set_title('Palabras más usadas cluster 3, k-means')

ax4.bar(df3km['Keys'], df3km['Values'],color=cividis.colors)
ax4.set_xticklabels(df3km['Keys'],rotation=45)
ax4.tick_params(axis='x', labelsize=n)
ax4.set_xlabel('Stemm')
ax4.set_ylabel('Frecuencia')
ax4.set_title('Palabras más usadas cluster 4, k-means')

#plt.tight_layout()
plt.savefig('kmeans.png')
```
